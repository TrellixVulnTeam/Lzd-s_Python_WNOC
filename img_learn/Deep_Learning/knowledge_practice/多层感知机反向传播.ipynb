```python
# 没有考虑到偏置项
import numpy as np
```

```python
# 数据个数,输入层神经元个数,隐含层神经元个数,输出层神经元个数
n,d_in,H,d_out = 64,1000,100,10
```

```python
x,y = np.random.randn(n,d_in),np.random.randn(n,d_out) # 生成原始数据
```

```python
w1,w2 = np.random.randn(d_in,H),np.random.randn(H,d_out) # 隐藏层和输出层的权重(输入层是没有权重的)
```

```python
w1.shape
```

```python
# 反向传播
for i in range(2000):
    # 隐含层的计算:采用tanh激活
    h_out = 1/(1+np.exp(-x.dot(w1))) # 隐含层输出64x100
    # 输出层:无激活函数
    y_pred = h_out.dot(w2) # 输出层64x10
    # 损失函数:mse
    loss = np.square(y_pred - y).sum()
    print(f"第{i+1}轮的损失函数为:{loss}")
    # 反向传播
    out_grad = 2*(y_pred - y) # 输出层的上游梯度64x10
    w2_grad = h_out.T.dot(out_grad) # 隐含层的上游梯度
    h_out_grad = out_grad.dot(w2.T)
    w1_grad = x.T.dot(h_out_grad*h_out*(1-h_out))
    
    w1 -= w1_grad*0.001
    w2 -= w2_grad*0.001
```
